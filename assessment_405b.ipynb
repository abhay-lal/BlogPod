```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import re
```

```python
df=pd.read_csv('assessment_405b.csv')
```

```python
df = df[df['id'] != 'id']
```

```python
df = df.drop_duplicates(subset='id')
```

```python
df 
```

```python
df[['Scores', 'Suggestions']] = df['assessment_405b'].str.split('\n\nSuggestions:', expand=True)
```

```python
scores_split = df['Scores'].str.split(',', expand=True)

scores_split = scores_split.reindex(columns=range(6))

scores_split.columns = ['Clarity', 'Grammar and Syntax', 'Tone Appropriateness', 
                        'Sentence Structure and Flow', 'Engagement', 'Conciseness']

df[['Clarity', 'Grammar and Syntax', 'Tone Appropriateness', 
    'Sentence Structure and Flow', 'Engagement', 'Conciseness']] = scores_split
```

```python
df[['Clarity', 'Grammar and Syntax', 'Tone Appropriateness', 
    'Sentence Structure and Flow', 'Engagement', 'Conciseness']].apply(pd.to_numeric, errors='coerce').hist(figsize=(12, 10))
```

```python

attributes = ["Clarity", "Grammar and Syntax", "Tone Appropriateness", 
              "Sentence Structure and Flow", "Engagement", "Conciseness"]

import re

def extract_attribute_comments_regex(suggestions_text, attribute, attributes_list):
    pattern = r'(\* \*\*' + re.escape(attribute) + r' \(\d+\):.*?)(?=\n\* \*\*(' + '|'.join([re.escape(attr) for attr in attributes_list]) + r') \(\d+\)|$)'
    
    match = re.search(pattern, suggestions_text, re.DOTALL)
    
    return match.group(1).strip() if match else None

for attribute in attributes:
    df[f"{attribute} Comments"] = df["Suggestions"].apply(lambda x: extract_attribute_comments_regex(str(x), attribute, attributes))


```

```python
df
```

```python
def clean_comments(comment, attribute):
    if pd.isna(comment):
        return comment
    return re.sub(r'\*\*' + re.escape(attribute) + r' \(\d+\):\*\* ', '', comment)

for attribute in attributes:
    df[f"{attribute} Comments"] = df[f"{attribute} Comments"].apply(lambda x: clean_comments(x, attribute))

df.head()
```

```python
import matplotlib.pyplot as plt

# Convert the scores columns to numeric
df[attributes] = df[attributes].apply(pd.to_numeric, errors='coerce')

# Plot the scores based on the ids
df.set_index('id')[attributes].plot(figsize=(15, 10), marker='o')
plt.title('Scores based on IDs')
plt.xlabel('ID')
plt.ylabel('Scores')
plt.legend(title='Attributes')
plt.grid(True)
plt.show()
```
